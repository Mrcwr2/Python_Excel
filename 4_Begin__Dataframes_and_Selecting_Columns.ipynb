### .loc[ ] and .iloc[ ] (Intermediate Topic)
---

.loc[] is a **name based operator**

.iloc[] is an index **position based operator**

<br>

**Example: Return one cell value**

    df.loc["row_name", "col_name"]


    df.iloc[row_num, col_num]

<br>


**Example: Return multiple values**

    df.loc[ ["row_name", "another_row_name"], ["col_name", "another_col_name"] ]


    df.iloc[ [row_num, another_row_num], [col_num, another_col_num] ]

## Common use of .loc[ ] - Slicing

    =SUM(A1:A10)

<br>

**:**  is the slice operator in Excel and in Pandas
* Used to select all rows/columns ( : )
* Used to select a range of rows/columns ("col1" : "col2")

```
# Select all rows and columns from "CERT" to "Acquiring Institution"
bank_name_index.loc[:,'CERT':'Acquiring Institution']
```
